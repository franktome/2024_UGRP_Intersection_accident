```python
import json
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
depth_path='all_depth_dic.json'
lane_path='x_dic.json'

with open(depth_path, 'r', encoding='utf-8') as json_file:
    depth = json.load(json_file)
with open(lane_path, 'r', encoding='utf-8') as json_file:
    lane = json.load(json_file)
```

```python
depth
```

```python
lane_data=lane
```

```python
depth
```

```python
import matplotlib.pyplot as plt

#  
plt.figure(figsize=(4, 8))  #  8,  8  
colors=['red', 'yellow', 'blue']
texts=[1, 1.5, 1.5]
count=0
for vehicle in lane_data.keys():
    #  
    y_values = depth[vehicle]  # y 
    x_values = lane_data[vehicle]  # x 

    #   
    plt.plot(x_values, y_values, label=vehicle, linewidth=2, color=colors[count])

    for i in range(0, len(x_values), 10):  # 10  
        alpha_value = 1 - (i / len(x_values))  # i   alpha 
        plt.scatter(x_values[i], y_values[i], color=colors[count], s=300, zorder=5, alpha=alpha_value)  #  
        
        #     (  )
        plt.text(x_values[i]+texts[count], y_values[i] + 1, str(round(i / 15, 1)), color='black', fontsize=10, ha='center', va='center', fontweight='bold')
    
    count += 1

#   
plt.legend(fontsize=12)  #  
plt.xlabel("x[m]", fontsize=14)
plt.ylabel("Depth[m]", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
